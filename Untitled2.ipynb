```python
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train / 255.0
x_test = x_test / 255.0
model = models.Sequential()
model.add(layers.Flatten(input_shape=(28, 28))) # 784차원 변환
model.add(layers.Dense(128, activation='relu')) # 은닉층 1 추가
model.add(layers.Dense(64, activation='relu'))  # 은닉층 2 추가
model.add(layers.Dense(10, activation='softmax')) # 출력층
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
history=model.fit(x_train, y_train, epochs=5)
test_loss, test_acc = model.evaluate(x_test, y_test)
print('테스트 정확도:', test_acc)
loss = history.history['loss']
acc = history.history['accuracy']
epochs = range(1, len(loss)+1)
plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, acc, 'r', label='Accuracy')
plt.xlabel('epochs')
plt.ylabel('loss/acc')
plt.show()
pred = model.predict(x_test[0].reshape(1,28,28), batch_size=1)
print("추정된 숫자=", pred.argmax())
plt.imshow(x_test[0], cmap="Greys")
```
